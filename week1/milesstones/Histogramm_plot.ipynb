### Step 3: Draw a histogram for 2015 with the default settings.

```python
import matplotlib.pyplot as plt
```

```python
lifeexp[2015].hist()
plt.savefig('histo.png')
```

### Step 4: Try a few values for the number of bins.

```python
lifeexp[2015].hist(bins=5)
plt.savefig('histo.png')
```

```python
lifeexp[2015].hist(bins=10)
plt.savefig('histo.png')
```

```python
lifeexp[2015].hist(bins=20)
plt.savefig('histo.png')
```

### Step 5: Modify the plot using some matplotlib functions.

```python
plt.title('life expactancy')
plt.xlabel('life expactancy in years')
plt.ylabel('relative frequency')
```

### Step 6: Set the min/max x and y values with plt.axis.

```python
lifeexp[2015].hist()
plt.title('Histogram')
plt.xlabel('life expactancy in years')
plt.ylabel('relative frequency')
plt.axis([0.0, 100.0, 0.0, 50.0])
plt.savefig('histo.png')
```

### Step 7: Try some optional parameters for the hist() function, e.g.:

```python
lifeexp[2015].hist(alpha=0.5, facecolor='#ff0000', histtype='bar')
plt.title('Histogram')
plt.xlabel('life expactancy in years')
plt.ylabel('relative frequency')
plt.axis([0.0, 100.0, 0.0, 50.0])

plt.savefig('histo.png')
```

### Step 8: Set the resolution of the image that is written.

```python
lifeexp[2015].hist()
plt.title('Histogram')
plt.xlabel('life expactancy in years')
plt.ylabel('relative frequency')
plt.axis([0.0, 100.0, 0.0, 50.0])
plt.savefig('histo.png', dpi=150)
```

#### write the diagram to a SVG file using the ending .svg

```python
lifeexp[2015].hist(alpha=0.5, facecolor='#ff0000', histtype='bar')
plt.title('Histogram')
plt.xlabel('life expactancy in years')
plt.ylabel('relative frequency')
plt.axis([0.0, 100.0, 0.0, 50.0]),
plt.savefig('histo.svg', dpi=150)
```
