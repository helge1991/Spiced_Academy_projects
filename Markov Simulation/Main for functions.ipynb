```python
#create crosstab with probabilities
P = pd.crosstab(index=df['location'],columns=df['next_location'],normalize='index').round(3)
P
```

```python
#check if sum = 1
states = df['location'].unique()
for state in states:
    print(f'Sum of Prob for {state} is {P.loc[state,:].sum()}')
```
